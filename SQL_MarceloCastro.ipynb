####SQL: Pregunta 1: Obtener los nombres de los alumnos que aprobaron, las materias y la nota, ordenado por la fecha del examen. Se aprueba la materia con una nota superior o igual a 7.


```python
# Ejercicio # 1
conn = sqlite3.connect('test.db')


####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''
SELECT e.nombre, m.descripcion ,e2.nota   
FROM Estudiante e
JOIN Examen e2 ON e2.id_estudiante  = e.id_estudiante 
JOIN Materia m ON m.id_materia = e2.id_materia 
WHERE e2.nota >= 7
ORDER BY e2.fecha ASC
 ;''')
for row in cursor:
  print(row)
conn.close()
```

####SQL: Pregunta 2: Obtener la mejor nota de cada nivel, por materia, incluir apellido, nombre del estudiante y materia.

```python
# Ejercicio # 2
conn = sqlite3.connect('test.db')


####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''
SELECT  m.nivel , m.id_materia, max(e.nota), e2.apellido , e2.nombre, m.descripcion 
FROM Materia m 
JOIN Examen e ON e.id_materia = m.id_materia 
JOIN Estudiante e2 ON e2.id_estudiante  = e.id_estudiante 
GROUP BY m.id_materia
ORDER BY m.nivel
 ;''')
for row in cursor:
  print(row)
conn.close()
```

####SQL: Pregunta 3: Obtener los estudiantes que hayan rendido mas de un examen de alguna materia siempre y cuando la segunda vez se haya obtenido una mejor nota, incluir la nota, el apellido y nombre del estudiante, la materia y el nivel.

```python
# Ejercicio # 3
conn = sqlite3.connect('test.db')


####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''
SELECT  nota2, e2.apellido ,e2.nombre, m.descripcion ,m.nivel
FROM Examen e
JOIN (
SELECT t2.id_estudiante as id_estudiante1 ,t2.id_materia as id_materia1 ,t2.fecha as fecha1, MIN(t2.nota) as nota1
FROM (
SELECT id_estudiante ,id_materia ,nota ,fecha 
FROM Examen e where id_estudiante in (SELECT id_estudiante 
FROM Examen e 
GROUP BY id_estudiante ,id_materia
HAVING COUNT(id_materia)>1) and id_materia  in (SELECT id_materia  
FROM Examen e 
GROUP BY id_estudiante ,id_materia
HAVING COUNT(id_materia)>1)
ORDER BY id_estudiante) as t2
GROUP BY t2.id_estudiante) AS a ON a.id_estudiante1 = e.id_estudiante 
JOIN (
SELECT t2.id_estudiante as id_estudiante2 ,t2.id_materia as id_materia2 ,t2.fecha as fecha2, MAX(t2.nota) as nota2
FROM (
SELECT id_estudiante ,id_materia ,nota ,fecha 
FROM Examen e where id_estudiante in (SELECT id_estudiante 
FROM Examen e 
GROUP BY id_estudiante ,id_materia
HAVING COUNT(id_materia)>1) and id_materia  in (SELECT id_materia  
FROM Examen e 
GROUP BY id_estudiante ,id_materia
HAVING COUNT(id_materia)>1)
ORDER BY id_estudiante) as t2
GROUP BY t2.id_estudiante) AS b ON b.id_estudiante2 = e.id_estudiante 
JOIN Estudiante e2 ON e2.id_estudiante = e.id_estudiante 
JOIN Materia m ON m.id_materia = e.id_materia
WHERE nota2 > nota1 and fecha2 < fecha1 
GROUP BY e.id_estudiante,e.id_materia  
HAVING COUNT(e.id_materia)>1
 ;''')
for row in cursor:
  print(row)
conn.close()
```

####SQL: Pregunta 4: Obtener los apellidos y nombres de los estudiantes que han rendido exámenes de materias de un nivel diferente al que están cursando, incluir la materia y la nota obtenida, ordenar alfabéticamente por el apellido del estudiante.

```python
# Ejercicio # 4
conn = sqlite3.connect('test.db')


####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''
SELECT e.apellido, e.nombre, m.descripcion,e2.nota  
FROM Estudiante e
JOIN Examen e2 ON e2.id_estudiante = e.id_estudiante 
JOIN Materia m ON m.id_materia  = e2.id_materia
WHERE e.nivel <> m.nivel 
ORDER BY e.apellido  
 ;''')
for row in cursor:
  print(row)
conn.close()
```

####SQL: Pregunta 5: Obtener los promedios totales por materia, en los casos de estudiantes que rindieron mas de un examen en alguna materia, considerar la mejor nota obtenida.

```python
# Ejercicio # 5
conn = sqlite3.connect('test.db')


####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''
SELECT m.descripcion, AVG(c.nota)
FROM (
SELECT id_estudiante, id_materia,  nota
FROM Examen e WHERE fecha  not in (
SELECT fecha 
FROM Examen e where id_estudiante in (SELECT id_estudiante 
FROM Examen e 
GROUP BY id_estudiante ,id_materia
HAVING COUNT(id_materia)>1) and id_materia  in (SELECT id_materia  
FROM Examen e 
GROUP BY id_estudiante ,id_materia
HAVING COUNT(id_materia)>1)
ORDER BY id_estudiante)
UNION ALL
SELECT id_estudiante, id_materia , MAX(nota) 
FROM Examen e where id_estudiante in (SELECT id_estudiante 
FROM Examen e 
GROUP BY id_estudiante ,id_materia
HAVING COUNT(id_materia)>1) and id_materia  in (SELECT id_materia  
FROM Examen e 
GROUP BY id_estudiante ,id_materia
HAVING COUNT(id_materia)>1)
GROUP BY  id_materia ) AS c 
JOIN Estudiante e2 ON e2.id_estudiante = c.id_estudiante 
JOIN Materia m ON m.id_materia = c.id_materia
GROUP BY c.id_materia
ORDER BY c.id_materia 
 ;''')
for row in cursor:
  print(row)
conn.close()
```
